```python
import pandas as pd
import pvdeg
```

```python
WEATHER_DB = "NSRDB"
WEATHER_ARG = {
    "satellite": "Americas",
    "names": "TMY",
    "NREL_HPC": True,
    "attributes": pvdeg.pysam.INSPIRE_NSRDB_ATTRIBUTES,
}

geo_weather, geo_meta = pvdeg.weather.get(
    WEATHER_DB, geospatial=True, **WEATHER_ARG
)
```

```python
assert geo_weather.attrs.get("kestrel_nsrdb_fnames") == ['/datasets/NSRDB/current/nsrdb_tmy-2024.h5']
```

```python
sherman_db_query = pd.read_csv('/home/tford/dev/InSPIRE/Studies/USMap_Doubleday_2024/nsrdb_goes_tmy_v4_0_0_202510061520.csv', index_col=0)
sherman_db_query.index.name = None
sherman_db_query
```

```python
geo_meta_overlapping = geo_meta.loc[sherman_db_query.index.values]
```

```python
pd.testing.assert_frame_equal(
    geo_meta_overlapping[["latitude", "longitude"]],
    sherman_db_query[["latitude", "longitude"]],
    check_dtype=False
)
```
