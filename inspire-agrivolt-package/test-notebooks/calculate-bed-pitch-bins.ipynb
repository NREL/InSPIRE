```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
CW = 2
TILT = 30 # latitude tilt but minimum is around 30 (south florida)

ground_pannel_proj = CW * np.cos(np.deg2rad(TILT)) # xp from spreadsheet
```

```python
pitches = np.linspace(3.8, 24, 1000)
```

```python
pitches
```

```python
edgetoedge = pitches - ground_pannel_proj
```

```python
edgetoedge
```

```python
under_pannel_percent = ground_pannel_proj * 100 / pitches
edgetoedge_percent = edgetoedge * 100 / pitches

edgetoedge_percent
```

```python
under_pannel_blocks = np.round(under_pannel_percent / 10)
under_pannel_blocks
```

```python
edgetoedge_blocks = np.round(edgetoedge_percent / 10)
edgetoedge_blocks
```

```python
shape = under_pannel_blocks
conf_array = np.ones_like(shape) * 10.0

# check that all under pannel counts + inter-row counts are the same
check_array = under_pannel_blocks + edgetoedge_blocks
np.testing.assert_array_equal(check_array, conf_array)
```

```python
outputs = []

prev_val = under_pannel_blocks[0]
for i in range(1, len(pitches)):
    cur_val = under_pannel_blocks[i]
    
    if prev_val != cur_val:
        outputs.append(i)

    prev_val = cur_val

outputs
```

```python
for n in pitches[outputs]:
    print(n)
```
